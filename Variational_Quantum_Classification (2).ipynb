# Variational Classification

Variational Classification is a technique within quantum machine learning that leverages quantum algorithms, specifically variational algorithms, to perform classification tasks. In this context, a classification task involves assigning input data into one of several predefined categories or classes.

***How Variational Classification Works***

1. **Parameterized Quantum Circuit (Ansatz):**
   - The core idea of variational classification is to use a parameterized quantum circuit, often called an ansatz, to process input data. This quantum circuit is designed with adjustable parameters that can be tuned to improve the classifier's performance.

2. **Data Encoding:**
   - The input data, which could be in the form of vectors (e.g., $\vec{x}_i$), is encoded into a quantum state using a data encoding unitary operation. This step prepares the quantum circuit to process the data in a quantum environment.

3. **Quantum State Preparation:**
   - After encoding the data, the quantum circuit generates a quantum state $ |\psi(\vec{x}_i; \vec{\theta}) \rangle $, where $\vec{\theta}$ represents the parameters of the circuit that need to be optimized. This state reflects the transformation of the initial data into a form that can be classified.

4. **Measurement and Output:**
   - The quantum state is then measured, and the result is typically a bitstring (a sequence of 0s and 1s). To classify the input data, this bitstring is processed through a Boolean function, often the parity function, which maps the bitstring to a binary output (e.g., 0 or 1 in binary classification).

5. **Training and Optimization:**
   - The parameters $\vec{\theta}$ of the quantum circuit are optimized during a training phase. This involves repeatedly running the quantum circuit, comparing the classifierâs predictions to the actual labels of the training data, and adjusting $\vec{\theta}$ to minimize a cost function that measures the accuracy of the classifier.

6. **Iterative Process:**
   - The optimization of the quantum circuitâs parameters is done iteratively, using a classical computer to evaluate the cost function and update the parameters. This process continues until the classifier achieves the desired level of accuracy.

***Key Characteristics***

- **Hybrid Quantum-Classical Approach:** Variational classification uses both quantum and classical computing resources, where quantum circuits handle the data encoding and transformation, while classical algorithms are used for optimization.

- **Near-Term Applicability:** Variational classification algorithms are designed to work with the current generation of quantum computers (Noisy Intermediate-Scale Quantum, or NISQ devices), which have limitations in terms of qubit count and coherence time.

- **Flexibility:** The approach can be adapted to different types of classification tasks, including binary and multi-class problems, by appropriately designing the quantum circuit and the data encoding scheme.

***Example: Variational Quantum Classifier (VQC)***

One popular implementation of variational classification is the Variational Quantum Classifier (VQC). In VQC, the quantum circuit is trained to perform binary classification by adjusting its parameters $\vec{\theta}$ so that the quantum state generated for each input data vector produces an accurate classification result. The training involves minimizing a cost function that reflects the difference between the predicted labels and the actual labels of the training dataset.

In summary, variational classification represents a promising approach to quantum machine learning, where quantum algorithms are utilized to classify data efficiently by leveraging the power of quantum circuits, optimized through classical computation.

# Variational Algorithms

Screenshot (345).png

Variational algorithms, which were first introduced in 2014, represent a significant advancement in the field of quantum computing. These algorithms were initially demonstrated through two foundational works: the **Variational Quantum Eigensolver (VQE)**, as detailed in Reference 1, and the **Quantum Approximate Optimization Algorithm (QAOA)**, as described in Reference 2. These early contributions have paved the way for a broader class of algorithms that are particularly well-suited for the current generation of quantum hardware. Variational algorithms are designed as [near-term algorithms](gloss:near-term-algorithm), meaning they are intended to be practical and effective on today's quantum computers, which are still limited in terms of qubit count and coherence times. To overcome these limitations, variational algorithms are executed in conjunction with classical computers, leveraging the strengths of both quantum and classical computing to solve complex problems.

At the heart of a variational algorithm is a parameterized quantum circuit, often referred to as an ansatz. This circuit is denoted as $ U(\theta) $, where $ \theta $ represents a set of parameters that can be adjusted during the algorithm's execution. The quantum circuit prepares a quantum state $ |\psi(\theta) \rangle = U(\theta) |0 \rangle $, where $ |0 \rangle $ is the initial state of the qubits, typically all set to zero. The quantum computer is then used to measure the expectation value of a particular observable, which is related to the problem being solved. This measurement process is a key step in evaluating how well the current parameters $ \theta $ are performing in terms of solving the given problem.

The performance of the parameters $ \theta $ is quantified using a cost function $ C(\theta) $. The cost function serves as a measure of how close the current quantum state $ |\psi(\theta) \rangle $ is to the desired solution. The goal of the algorithm is to find the optimal set of parameters $ \theta $ that minimize this cost function. This optimization process is iterative, where a classical computer plays a crucial role. After each iteration, the classical computer calculates the value of the cost function based on the measurements obtained from the quantum computer. It then uses an optimization algorithm, such as gradient descent or another suitable method, to adjust the parameters $ \theta $ in a way that reduces the cost function. This feedback loop between the quantum and classical computers continues until the cost function is minimized, leading to the identification of the optimal circuit parameters $ \theta $.

In summary, variational algorithms are a powerful approach in quantum computing that harnesses the strengths of both quantum and classical systems. By iteratively refining the parameters of a quantum circuit to minimize a cost function, these algorithms aim to solve complex problems efficiently, even with the current limitations of quantum hardware. The introduction of the **Variational Quantum Eigensolver** and the **Quantum Approximate Optimization Algorithm** in 2014 marked the beginning of this promising area of research, which continues to evolve and expand as quantum technology advances.

# Variational Quantum Classifier

The Variational Quantum Classifier (VQC) is a specialized type of variational algorithm, first introduced by multiple research groups in 2018, that extends the principles of variational algorithms to the domain of machine learning. Specifically, the VQC is designed for classification tasks, where the goal is to assign input data to one of several predefined categories. In the context of a binary classification problem, where each input is associated with one of two possible output labels (typically represented as 0 or 1), the VQC interprets the measured expectation value from a quantum circuit as the output of a classifier.

For a binary classification task, we begin with a set of input data vectors $ \vec{x}_i $, where each vector corresponds to a particular instance in the dataset, and each instance has an associated binary label $ y_i \in \{0, 1\} $. The VQC framework involves building a parameterized quantum circuit for each input data vector. The circuit prepares a quantum state according to the following equation:

$$ \cssId{_eq_1}{|\psi(\vec{x}_i;\vec{\theta}) \rangle =  U_{W(\vec{\theta})}U_{\phi(\vec{x}_i)}|0 \rangle}$$

In this equation, $ U_{W(\vec{\theta})} $ represents the unitary operation associated with the variational circuit, which is parameterized by a set of variables $ \vec{\theta} $. These parameters $ \vec{\theta} $ are adjusted during the training process to optimize the classifier's performance. The term $ U_{\phi(\vec{x}_i)} $ denotes the unitary operation that encodes the input data vector $ \vec{x}_i $ into the quantum state. Together, these operations transform the initial quantum state $ |0 \rangle $ (usually representing all qubits in the ground state) into a quantum state $ |\psi(\vec{x}_i;\vec{\theta}) \rangle $ that encodes the information necessary for classification.

Once the circuit is constructed and the quantum state $ |\psi(\vec{x}_i;\vec{\theta}) \rangle $ is prepared, we proceed to measure the quantum state. The measurement is performed across all $ n $ qubits, yielding a bitstring of length $ n $. This bitstring is a classical representation of the quantum state and serves as the basis for determining the classifier's output. To derive the binary output label (either 0 or 1) from this bitstring, we apply a Boolean function $ f: \{0, 1\}^{n} \rightarrow \{0, 1\} $. This function maps the bitstring to a single binary value, which constitutes the classifier's prediction for that particular input. The [parity function](gloss:parity-function), which computes the parity (even or odd sum) of the bits in the bitstring, is a commonly used choice for this mapping due to its simplicity and effectiveness in distinguishing between the two possible classes.

The VQC operates within a training phase where the objective is to optimize the parameters $ \vec{\theta} $ so that the quantum circuit produces the most accurate classifications. During training, the quantum classifier generates predicted labels $ \hat{y_i} $ for each input data vector. These predicted labels are then compared to the actual labels $ y_i $ provided in the training dataset. The discrepancy between the predicted and actual labels is quantified using a cost function, which serves as a measure of the classifier's performance. A lower cost indicates better alignment between predictions and actual labels.

Based on the value of the cost function, a classical optimization algorithm is employed to adjust the parameters $ \vec{\theta} $. This algorithm iteratively updates the parameters in a manner that seeks to minimize the cost function, thereby improving the classifier's accuracy. Once the parameters have been updated, a new quantum circuit is executed with the revised parameters, and the process is repeated. This iterative loop continues until the cost function stabilizes, indicating that the classifier has reached an optimal set of parameters $ \vec{\theta} $ and is ready to be used for making accurate predictions on new, unseen data.

In essence, the Variational Quantum Classifier represents a powerful approach that leverages both quantum and classical computing resources to tackle machine learning problems, particularly in scenarios where traditional classifiers may struggle to find optimal solutions. By combining the expressive power of quantum circuits with the optimization capabilities of classical algorithms, the VQC offers a promising avenue for advancing the field of quantum machine learning.

This function ensures that the correct values are assigned to the corresponding quantum gates, thereby enabling the circuit to accurately encode the input data and apply the appropriate transformations during the classification process.

### Building Ad Hoc Circuits

```python
def circuit_instance(data, variational):
    """Assigns parameter values to `AD_HOC_CIRCUIT`.
    Args:
        data (list): Data values for the feature map
        variational (list): Parameter values for `VAR_FORM`
    Returns:
        QuantumCircuit: `AD_HOC_CIRCUIT` with parameters assigned
    """
    # pylint: disable=invalid-name
    parameters = {}
    for i, p in enumerate(FEATURE_MAP.ordered_parameters):
        parameters[p] = data[i]

    for i, p in enumerate(VAR_FORM.ordered_parameters):
        parameters[p] = variational[i]

    return AD_HOC_CIRCUIT.assign_parameters(parameters)
```

## Step 4

To finalize the classification process in the variational quantum classifier, we need to create a class assignment function that determines the output label based on the parity of the measured bitstring. After the quantum circuit processes the input data and the qubits are measured, the result is a bitstring composed of 0s and 1s.

This function calculates the parity of the bitstring, which is the sum of the bits modulo 2. The parity is considered even if the sum of the bits is even (e.g., 0, 2, 4, etc.), and odd if the sum is odd (e.g., 1, 3, 5, etc.). Based on this parity, the function assigns a label: if the parity is even, the function returns a label of 1; if the parity is odd, it returns a label of 0. This simple but effective method of class assignment allows the quantum classifier to translate the quantum measurement outcomes into meaningful predictions.

```python
def parity(bitstring):
    """Returns 1 if parity of `bitstring` is even, otherwise 0."""
    hamming_weight = sum(int(k) for k in list(bitstring))

    return (hamming_weight+1) % 2
```

## Step 5

To interpret the results from running a quantum circuit multiple times, we create a function that returns the probability distribution over the label classes based on the experimental counts. When a quantum circuit is executed on a quantum device or simulator, it typically runs for many iterations (shots), and the result is a collection of bitstrings with associated counts indicating how often each bitstring was observed.

This function takes these counts as input and calculates the probability of each label class by summing the counts of bitstrings that correspond to each class. Specifically, for each unique bitstring, the function determines the class label by evaluating the parity (as previously defined) and then calculates the probability as the ratio of the count for that label to the total number of shots. This gives a distribution over the label classes, reflecting the likelihood that a given input will be classified into each category based on the quantum circuit's outputs.

```python
def label_probability(results):
    """Converts a dict of bitstrings and their counts, to parities and their counts"""

    shots = sum(results.values())
    probabilities = {0: 0, 1: 0}

    for bitstring, counts in results.items():
        label = parity(bitstring)
        probabilities[label] += counts / shots

    return probabilities
```

## Step 6

To classify data using our variational quantum classifier, we create a function that orchestrates the entire classification process. This function takes as input the dataset and the variational parameters that define the quantum circuit's behavior.

For each data point in the dataset, the function first assigns the specific data values to the feature map, ensuring that the quantum circuit accurately encodes the input information into quantum states. Concurrently, the variational parameters are assigned to the corresponding gates in the variational circuit, setting up the quantum operations that will be optimized during training.

### Classification Probabilities

```python
from qiskit_aer import Aer
from qiskit.primitives import Sampler
```

```python
def classification_probability(data, variational):
    """Classify data points using given parameters.
    Args:
        data (list): Set of data points to classify
        variational (list): Parameters for `VAR_FORM`
    Returns:
        list[dict]: Probability of circuit classifying
                    each data point as 0 or 1.
    """
    circuits = [circuit_instance(d, variational) for d in data]
    backend = Aer.get_backend('qasm_simulator')
    sampler = Sampler()
    results = sampler.run(circuits, backend).result()

    classification = [label_probability(results.get_counts(c)) for c in circuits]

    return classification
```

## Step 7

### Cross Entropy Loss Funtion

```python
def cross_entropy_loss(classification, expected):
    """Calculate accuracy of predictions using cross entropy loss.
    Args:
        classification (dict): Dict where keys are possible classes,
                               and values are the probability our
                               circuit chooses that class.
        expected (int): Correct classification of the data point.

    Returns:
        float: Cross entropy loss
    """
    # pylint: disable=invalid-name
    p = classification.get(expected)  # Prob. of correct classification
    return -np.log(p + 1e-10)
```

### Cost Function

```python
def cost_function(data, labels, variational):
    """Evaluates performance of our circuit with `variational`
    parameters on `data`.

    Args:
        data (list): List of data points to classify
        labels (list): List of correct labels for each data point
        variational (list): Parameters to use in circuit

    Returns:
        float: Cost (metric of performance)
    """
    # pylint: disable=invalid-name
    classifications = classification_probability(data, variational)
    cost = 0

    for i, classification in enumerate(classifications):
        cost += cross_entropy_loss(classification, labels[i])

    cost /= len(data)

    return cost
```

## Step 8

To train our variational quantum classifier, we need to set up a classical optimizer that will iteratively adjust the parameters of our variational circuit to minimize the cost function. For this task, we use the Simultaneous Perturbation Stochastic Approximation (SPSA) optimizer, which is well-suited for optimizing noisy or complex functions like those encountered in quantum machine learning. SPSA is an efficient gradient-free optimization method that requires only two measurements per iteration, making it ideal for variational quantum algorithms where function evaluations can be expensive.

We begin by initializing the variational circuit parameters to ensure reproducibility. This step is critical because quantum optimization can be sensitive to the initial parameter values, and setting a fixed seed allows us to achieve consistent results across different runs.

With the SPSA optimizer configured, we then proceed to optimize the cost function using our 40 training data points. The cost function quantifies the difference between the predicted labels from the quantum classifier and the actual labels from the training dataset. During the optimization process, SPSA adjusts the variational parameters in small steps, guided by the cost functionâs feedback. Each iteration of the optimization involves running the quantum circuits, measuring the outcomes, and updating the parameters to reduce the cost.

This optimization process is computationally intensive and may take some time to complete, especially when working with a large number of training data points or a complex quantum circuit. However, once the optimization is complete, the variational circuit should be well-tuned to classify new data points with improved accuracy.

### Initiating the Optimizer

```python
class OptimizerLog:  # pylint: disable=too-few-public-methods
    """Log to store optimizer's intermediate results"""
    def __init__(self):
        self.evaluations = []
        self.parameters = []
        self.costs = []

    def update(self, evaluation, parameter, cost, _stepsize, _accept):
        """Save intermediate results. Optimizer passes five values
        but we ignore the last two."""
        self.evaluations.append(evaluation)
        self.parameters.append(parameter)
        self.costs.append(cost)
```

### Objective Function

```python
def objective_function(variational):
    """Cost function of circuit parameters on training data.
    The optimizer will attempt to minimize this."""
    if not isinstance(variational, np.ndarray):
        variational = np.array(variational)
    if len(variational.shape) == 1:
        variational = variational.reshape(1, -1)

    return cost_function(TRAIN_DATA, TRAIN_LABELS, variational)
```

```python
def classification_probability(data, variational):
    circuits = [circuit_instance(d, variational) for d in data]
    backend = Aer.get_backend('qasm_simulator')
    sampler = Sampler()

    # Ensure parameter_values is a sequence of sequences of floats
    parameter_values = [
        [float(p) for p in circuit.parameters]
        for circuit in circuits
    ]

    results = sampler.run(circuits, parameter_values=parameter_values, backend=backend).result()

    classification = [label_probability(results.quasi_dists[i]) for i in range(len(circuits))]

    return np.array(classification)
```

## Step 9

To evaluate the performance of our trained variational quantum classifier, we implement a scoring function that leverages the classification function we previously developed. This scoring function will assess how well our classifier performs on unseen dataâin this case, our 10 test data points.

The function first takes in the test dataset and the optimized variational parameters obtained from the training process. For each test data point, the function uses the classification function to predict the class label by running the quantum circuit and analyzing the measurement outcomes.

Once all predictions are made, the function compares them to the actual labels of the test data points. The score is then calculated as the proportion of correctly classified data points out of the total number of test points. This accuracy score provides a quantitative measure of the classifierâs performance, indicating how well the quantum classifier can generalize to new, unseen data after being trained on the initial dataset.

### Test Classifier

```python
def test_classifier(data, labels, variational):
    """Gets classifier's most likely predictions and accuracy of those
    predictions.

    Args:
        data (list): List of data points to classify
        labels (list): List of correct labels for each data point
        variational (list): List of parameter values for classifier

    Returns:
        float: Average accuracy of classifier over `data`
        list: Classifier's label predictions for each data point
    """
    probability = classification_probability(data, variational)
    predictions = [0 if p[0] >= p[1] else 1 for p in probability]
    accuracy = 0
    # pylint: disable=invalid-name
    for i, prediction in enumerate(predictions):
        if prediction == labels[i]:
            accuracy += 1
    accuracy /= len(labels)
    return accuracy, predictions
```

# Variational Training

When working with variational algorithms, one of the most time-consuming and computationally intensive tasks is finding the optimal parameters for the variational circuit. This process is heavily dependent on the optimization method chosen, as discussed in the [training page](./training-quantum-circuits).

Screenshot (347).png

The goal is to find the optimal set of circuit parameters, denoted as $\vec{\theta}^*$, by minimizing a loss function $f(\vec{\theta})$. However, the relationship between the loss function and the circuit parameters is typically complex and non-linear.

The resulting loss landscape can be intricate, characterized by numerous hills and valleys, making the optimization process challenging. During optimization, the algorithm navigates this landscape, seeking the global minimum where the loss function is minimized. However, as illustrated by the black points and lines in the example below, about two-thirds of the optimization attempts may end up in local minima instead of finding the global minimum. This outcome is a common challenge in variational algorithms, where the complexity of the loss landscape can trap the optimization process in suboptimal solutions.

Screenshot (348).png

Optimization methods for variational algorithms can generally be divided into two categories: gradient-based and gradient-free methods. Gradient-based methods search for an optimal solution by identifying a point where the gradient of the loss function is zero, indicating an extremum. The direction of the search is guided by the derivative of the loss function. However, these methods have significant drawbacks, including potentially slow convergence and the risk of getting stuck in local optima without reaching the global solution.

On the other hand, when derivative information is unavailable or difficult to obtainâsuch as when the loss function is computationally expensive to evaluate or noisyâgradient-free methods offer a valuable alternative. These methods are robust in finding global optima, whereas gradient-based approaches often converge to local optima. The trade-off, however, is that gradient-free methods typically demand more computational resources, especially in problems with high-dimensional search spaces.

Screenshot (349).png

Regardless of the optimization method used, a common challenge arises when the loss landscape is relatively flat, making it difficult for the optimizer to determine an effective search direction. This situation, known as a _[barren plateau](gloss:barren-plateaus),_ was explored in depth in Reference 4. For many reasonable parameterized quantum circuits, the likelihood of obtaining a non-zero gradient along any given direction decreases exponentially with the number of qubits, making it increasingly difficult to optimize the circuit as the system size grows.

To address the barren plateau problem, researchers have proposed several strategies. One approach involves using structured initial guesses, similar to those used in quantum simulation, to provide the optimizer with a better starting point. Another potential solution is to design the quantum circuit as a sequence of shallow blocks, randomly selecting some parameters and setting the others such that each block implements the identity operation. This technique reduces the effective depth of the circuit and can mitigate the barren plateau issue. These strategies are still under active investigation, as they represent a promising direction for improving the performance of variational quantum algorithms.
